```python
import json
import csv
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_json('Appliances_5.json', lines=True)
df.to_csv('appliances.csv')
```

```python
s_df = pd.read_csv("appliances.csv")
```

```python
s_df.head()
```

```python
# Adding the Qunatity to the given database to make the transactinal database.
s_df['quantity'] = 1
```

```python
#Group by the data based on the customer (reviewer ID)
print(len(s_df['reviewerID'].unique()))
s = s_df.groupby('reviewerID')
s.first()
```

```python
# This basket data will contain the Quantity of each items bought per transaction (Reviewer ID and asin)
basket_plus = (s_df.groupby(['reviewerID', 'asin'])['quantity']
               .sum().unstack().reset_index().fillna(0).set_index('reviewerID'))
basket_plus
```

```python
#To encode the basket data into a binary data that shows whether an items is bought (1) or not (0).
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
basket_encode_plus = basket_plus.applymap(encode_units)
basket_encode_plus
```

```python
# finding the customer that has bought more than 2 items.

basket_filter_plus = basket_encode_plus[(basket_encode_plus > 0).sum(axis=1) >= 2]
basket_filter_plus
```

```python
#Finding the frequent itemsets using apriori algorithm based on the support value.

frequent_itemsets_plus = apriori(basket_filter_plus, min_support=0.04, 
                                 use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus
```

```python
# finding the frequent itemsets based on the length and support

frequent_itemsets_plus[ (frequent_itemsets_plus['length']  >= 2) &
                        (frequent_itemsets_plus['support'] >= 0.04) ]
```

```python
#Finding the association rules based on the lift matric
association_rules(frequent_itemsets_plus, metric='lift', 
                  min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```
